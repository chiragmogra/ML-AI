```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df=pd.read_csv('http://13.234.66.67/summer19/datasets/bank.csv')
```

```
plt.bar(df['Gender'],df['Balance'])
plt.xlabel('Gender')
plt.ylabel('Balance')
plt.title('bar between Gender and Balance')
plt.show()
```

```
plt.bar(df['Gender'],df['EstimatedSalary'])
plt.xlabel('Gender')
plt.ylabel('Estimated Salary')
plt.title('bar between Gender and Estimated Salary')
plt.show()
```

```
plt.bar(df['Age'],df['Balance'])
plt.xlabel('Age')
plt.ylabel('Balance')
plt.title('bar between age and Balance')
plt.show()
```
